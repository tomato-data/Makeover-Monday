```python
# code for letting the plots not to disappear
import plotly.io as pio
pio.renderers.default='notebook'

import pandas as pd

df_energy = pd.read_csv('global energy consumption.csv')
for i in range(3, 12):
    df_energy.iloc[:, i+1] += df_energy.iloc[:, i]
df_energy
```

```python
df_energy.columns
```

```python
import plotly.graph_objs as go

fig = go.Figure()
fig.add_trace(go.Scatter(
name = 'Traditional biomass',
x = df_energy['Year'],
y = df_energy['Traditional biomass'],
fill = 'tozeroy',
mode = 'lines'))

for en in df_energy.columns[4:]:
    fig.add_trace(go.Scatter(
    name = en,
    x = df_energy['Year'],
    y = df_energy[en],
    fill = 'tonexty',
    mode = 'lines'))
    
fig.update_layout(title = 'Global primary energy consumption by source',
                  yaxis=dict(range=[0, 190000], ticksuffix = 'TWh', tickformat=',', title='Energy Consumption (TWh)'),
                  paper_bgcolor='darkgray',
                  plot_bgcolor='darkgray')

fig.show()
```

```python
df_ren_energy = pd.read_csv('global renewable energy.csv')
for i in range(3, 8):
    df_ren_energy.iloc[:, i+1] += df_ren_energy.iloc[:, i]
df_ren_energy
```

```python
import plotly.graph_objs as go

fig = go.Figure()
fig.add_trace(go.Scatter(
name = 'Hydropower',
x = df_ren_energy['Year'][17:],
y = df_ren_energy['Hydropower'],
fill = 'tozeroy',
mode = 'lines'))

for ren in df_ren_energy.columns[4:]:
    fig.add_trace(go.Scatter(
    name = ren,
    x = df_ren_energy['Year'][17:],
    y = df_ren_energy[ren],
    fill = 'tonexty',
    mode = 'lines'))
    
fig.update_layout(title = 'Global primary energy consumption by source',
                  yaxis=dict(range=[0, 19000], ticksuffix = 'TWh', tickformat=',', title='Energy Consumption (TWh)'),
                  paper_bgcolor='darkgray',
                  plot_bgcolor='darkgray')
    
fig.show()
```

```python
# sum 사용법
df_energy2 = pd.read_csv('global energy consumption.csv')
sum_row = df_energy2[[x for x in df_energy.columns[3:]]].sum()

df_energy2.loc['Total'] = sum_row
df_energy2.loc['Total', 'Year'] = 'Total'
df_energy2
```

```python
df_energy2 = pd.read_csv('global energy consumption.csv')
for i in range(1, df_energy2.shape[0]):
    df_energy2.iloc[i, 3:] += df_energy2.iloc[i-1, 3:]

df_energy2
```

```python
df_energy2 = pd.read_csv('global energy consumption.csv')
# Calculate cumulative sum for columns starting from the 4th column (index 3)
df_energy2.iloc[:, 3:] = df_energy2.iloc[:, 3:].cumsum()

df_energy2
```

```python
fig = go.Figure()

energy = list(df_energy2.columns[3:])
for en in energy:
    fig.add_trace(go.Scatter(
    name = en,
    x = df_energy2['Year'][17:],
    y = df_energy2[en][17:],
    mode = 'lines'))
    
fig.update_layout(
    title = 'Total energy consumption accumulated by energy',
    xaxis_title='Year',
    yaxis_title='Total Energy Accumulated',
    paper_bgcolor='darkgray',
    plot_bgcolor='darkgray')
fig.show()
```

```python
fig = go.Figure()

energy = ['Traditional biomass', 'Coal', 'Oil', 'Gas']
for en in energy:
    fig.add_trace(go.Scatter(
    name = en,
    x = df_energy2['Year'][17:54],
    y = df_energy2[en][17:54],
    mode = 'lines'))
    
fig.add_shape(
    type="line",
    x0=1968.5, y0=0, x1=1968.5, y1=1,
    xref='x', yref='paper',  # 'paper' value for yref makes y-axis scale from 0 to 1
    line=dict(color="Red", width=2, dash="dash")
)
fig.add_shape(
    type="line",
    x0=1971, y0=0, x1=1971, y1=1,
    xref='x', yref='paper',  # 'paper' value for yref makes y-axis scale from 0 to 1
    line=dict(color="Green", width=2, dash="dash")
)
fig.add_shape(
    type="line",
    x0=1972.5, y0=0, x1=1972.5, y1=1,
    xref='x', yref='paper',  # 'paper' value for yref makes y-axis scale from 0 to 1
    line=dict(color="Green", width=2, dash="dash")
)
fig.add_shape(
    type="line",
    x0=1993.5, y0=0, x1=1993.5, y1=1,
    xref='x', yref='paper',  # 'paper' value for yref makes y-axis scale from 0 to 1
    line=dict(color="Purple", width=2, dash="dash")
)

fig.add_annotation(
    x=1968.5, y=1.05,  # Adjust y position to move the annotation slightly above the top of the plot
    text="TB < Coal",
    showarrow=False,
    arrowhead=2,
    ax=0, ay=-40,  # Set a negative value for ay to position the arrow above the annotation
    xref="x", yref="paper",  # Set the reference of the annotation to the x-axis and the entire plot area
    align="left"  # Align the annotation to the left of the specified x-coordinate
)
fig.add_annotation(
    x=1971, y=1.1, 
    text="TB < Oil",
    showarrow=False,
    arrowhead=2,
    ax=0, ay=-40,
    xref="x", yref="paper",  
    align="left" 
)
fig.add_annotation(
    x=1972.5, y=1.05, 
    text="Coal < Oil",
    showarrow=False,
    arrowhead=2,
    ax=0, ay=-40,
    xref="x", yref="paper",  
    align="left" 
)
fig.add_annotation(
    x=1993.5, y=1.05, 
    text="TB < Gas",
    showarrow=False,
    arrowhead=2,
    ax=0, ay=-40,
    xref="x", yref="paper",  
    align="left" 
)
    
fig.update_layout(
    title = 'Total energy consumption accumulated by energy source',
    xaxis_title='Year',
    yaxis_title='Total Energy Accumulated',
    paper_bgcolor='darkgray',
    plot_bgcolor='darkgray')
fig.show()
```

```python
def compare_consumption(low_en, high_en):
    for index, row in df_energy2.iterrows():
        if row[low_en] < row[high_en]:
            return row['Year']
    return None
print(compare_consumption('Traditional biomass', 'Gas'))
```
